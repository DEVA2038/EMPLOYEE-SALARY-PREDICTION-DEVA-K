```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['workclass']=encoder.fit_transform(data['workclass'])
data['marital-status']=encoder.fit_transform(data['marital-status'])
data['occupation']=encoder.fit_transform(data['occupation'])
data['relationship']=encoder.fit_transform(data['relationship'])
data['race']=encoder.fit_transform(data['race'])
data['gender']=encoder.fit_transform(data['gender'])
data['native-country']=encoder.fit_transform(data['native-country'])
data
```

```python
x=data.drop(columns=['income'])
y=data['income']
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, classification_report
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
xtrain, xtest, ytrain, ytest = train_test_split(
    x_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
xtrain
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(xtrain, ytrain)
pred_knn = knn.predict(xtest)

print("🔹 KNN Accuracy:", accuracy_score(ytest, pred_knn))
print(classification_report(ytest, pred_knn))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(xtrain, ytrain)
pred_lr = lr.predict(xtest)

print("🔹 Logistic Regression Accuracy:", accuracy_score(ytest, pred_lr))
print(classification_report(ytest, pred_lr))
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(solver="adam", hidden_layer_sizes=(5, 2), random_state=2, max_iter=2000)
mlp.fit(xtrain, ytrain)
pred_mlp = mlp.predict(xtest)

print("🔹 MLP Neural Network Accuracy:", accuracy_score(ytest, pred_mlp))
print(classification_report(ytest, pred_mlp))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(xtrain, ytrain)
pred_dt = dt.predict(xtest)

print("🔹 Decision Tree Accuracy:", accuracy_score(ytest, pred_dt))
print(classification_report(ytest, pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(xtrain, ytrain)
pred_rf = rf.predict(xtest)

print("🔹 Random Forest Accuracy:", accuracy_score(ytest, pred_rf))
print(classification_report(ytest, pred_rf))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier()
gb.fit(xtrain, ytrain)
pred_gb = gb.predict(xtest)

print("🔹 Gradient Boosting Accuracy:", accuracy_score(ytest, pred_gb))
print(classification_report(ytest, pred_gb))
```

```python
from sklearn.svm import SVC

svm = SVC()
svm.fit(xtrain, ytrain)
pred_svm = svm.predict(xtest)

print("🔹 SVM Accuracy:", accuracy_score(ytest, pred_svm))
print(classification_report(ytest, pred_svm))
```

```python
results = {
    "KNN": accuracy_score(ytest, pred_knn),
    "Logistic Regression": accuracy_score(ytest, pred_lr),
    "MLP Neural Net": accuracy_score(ytest, pred_mlp),
    "Decision Tree": accuracy_score(ytest, pred_dt),
    "Random Forest": accuracy_score(ytest, pred_rf),
    "Gradient Boosting": accuracy_score(ytest, pred_gb),
    "SVM": accuracy_score(ytest, pred_svm)
}
```

```python
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.bar(results.keys(), results.values(), color='royalblue')
plt.ylabel('Accuracy Score')
plt.title('Model Accuracy Comparison')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import joblib
model_objects = {
    "KNN": knn,
    "Logistic Regression": lr,
    "MLP Neural Net": mlp,
    "Decision Tree": dt,
    "Random Forest": rf,
    "Gradient Boosting": gb,
    "SVM": svm
}

sorted_results = sorted(results.items(), key=lambda item: item[1], reverse=True)
print("\n📊 Model Ranking by Accuracy:")
for rank, (model_name, score) in enumerate(sorted_results, 1):
    print(f"{rank}. {model_name}: {score:.4f}")
best_model = sorted_results[0]
print(f"\n🏆 Most Efficient Model: {best_model[0]} with Accuracy: {best_model[1]:.4f}")
best_model_name, best_accuracy = sorted_results[0]
best_model = model_objects[best_model_name]
joblib.dump(best_model, "best_model.pkl")
print("✅ Best model saved as best_model.pkl")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
import plotly.express as px

# Page config
st.set_page_config(
    page_title="💰 AI Salary Predictor",
    page_icon="💰",
    layout="wide",
    initial_sidebar_state="expanded"
)

# Label encoders (must match training)
workclass_map = {
    "Private": 0, "Self-emp-not-inc": 1, "Self-emp-inc": 2,
    "Federal-gov": 3, "Local-gov": 4, "State-gov": 5,
    "Without-pay": 6, "Never-worked": 7, "Unknown": 8
}

marital_map = {
    "Married-civ-spouse": 0, "Divorced": 1, "Never-married": 2,
    "Separated": 3, "Widowed": 4, "Married-spouse-absent": 5,
    "Married-AF-spouse": 6, "Unknown": 2
}

occupation_map = {
    "Tech-support": 0, "Craft-repair": 1, "Other-service": 2,
    "Sales": 3, "Exec-managerial": 4, "Prof-specialty": 5,
    "Handlers-cleaners": 6, "Machine-op-inspct": 7, "Adm-clerical": 8,
    "Farming-fishing": 9, "Transport-moving": 10, "Priv-house-serv": 11,
    "Protective-serv": 12, "Armed-Forces": 13, "Unknown": 2
}

relationship_map = {
    "Wife": 0, "Own-child": 1, "Husband": 2,
    "Not-in-family": 3, "Other-relative": 4, "Unmarried": 5,
    "Unknown": 3
}

race_map = {
    "White": 0, "Asian-Pac-Islander": 1, "Amer-Indian-Eskimo": 2,
    "Other": 3, "Black": 4, "Unknown": 0
}

gender_map = {
    "Female": 0, "Male": 1, "Unknown": 0
}

country_map = {
    "United-States": 0, "Cambodia": 1, "England": 2, "Puerto-Rico": 3,
    "Canada": 4, "Germany": 5, "Outlying-US(Guam-USVI-etc)": 6,
    "India": 7, "Japan": 8, "Greece": 9, "South": 10, "China": 11,
    "Cuba": 12, "Iran": 13, "Honduras": 14, "Philippines": 15,
    "Italy": 16, "Poland": 17, "Jamaica": 18, "Vietnam": 19,
    "Mexico": 20, "Portugal": 21, "Ireland": 22, "France": 23,
    "Dominican-Republic": 24, "Laos": 25, "Ecuador": 26, "Taiwan": 27,
    "Haiti": 28, "Columbia": 29, "Hungary": 30, "Guatemala": 31,
    "Nicaragua": 32, "Scotland": 33, "Thailand": 34, "Yugoslavia": 35,
    "El-Salvador": 36, "Trinadad&Tobago": 37, "Peru": 38, "Hong": 39,
    "Holand-Netherlands": 40, "Unknown": 0
}

@st.cache_resource
def load_model():
    try:
        return joblib.load("/content/best_model.pkl")
    except Exception as e:
        st.error(f"Model Load Error: {e}")
        return None

model = load_model()

def preprocess_input(input_dict):
    df = pd.DataFrame([input_dict])
    df['workclass'] = df['workclass'].map(workclass_map).fillna(8)
    df['marital-status'] = df['marital-status'].map(marital_map).fillna(2)
    df['occupation'] = df['occupation'].map(occupation_map).fillna(2)
    df['relationship'] = df['relationship'].map(relationship_map).fillna(3)
    df['race'] = df['race'].map(race_map).fillna(0)
    df['gender'] = df['gender'].map(gender_map).fillna(0)
    df['native-country'] = df['native-country'].map(country_map).fillna(0)

    expected_columns = [
        'age', 'workclass', 'fnlwgt', 'educational-num',
        'marital-status', 'occupation', 'relationship', 'race',
        'gender', 'capital-gain', 'capital-loss', 'hours-per-week',
        'native-country'
    ]
    return df[expected_columns]

st.markdown("## 💰 AI-Powered Income Classification System")
st.markdown("Predict whether income exceeds $50K/year based on individual attributes.")

if model is None:
    st.stop()

# Tabs
tab1, tab2, tab3 = st.tabs(["🔮 Individual Prediction", "📊 Batch Analysis", "ℹ️ Model Info"])

with tab1:
    st.markdown("### 👤 Individual Details")
    col1, col2 = st.columns(2)

    with col1:
        age = st.slider("Age", 17, 90, 35)
        fnlwgt = st.number_input("Final Weight", min_value=10000, max_value=1500000, value=200000)
        education_num = st.slider("Education Level (1-16)", 1, 16, 9)
        workclass = st.selectbox("Work Class", sorted(workclass_map.keys()))
        marital_status = st.selectbox("Marital Status", sorted(marital_map.keys()))
        occupation = st.selectbox("Occupation", sorted(occupation_map.keys()))
        relationship = st.selectbox("Relationship", sorted(relationship_map.keys()))

    with col2:
        race = st.selectbox("Race", sorted(race_map.keys()))
        gender = st.selectbox("Gender", sorted(gender_map.keys()))
        capital_gain = st.number_input("Capital Gain", 0, 100000, 0)
        capital_loss = st.number_input("Capital Loss", 0, 5000, 0)
        hours_per_week = st.slider("Hours per Week", 1, 99, 40)
        native_country = st.selectbox("Native Country", sorted(country_map.keys()))

    input_data = {
        'age': age,
        'workclass': workclass,
        'fnlwgt': fnlwgt,
        'educational-num': education_num,
        'marital-status': marital_status,
        'occupation': occupation,
        'relationship': relationship,
        'race': race,
        'gender': gender,
        'capital-gain': capital_gain,
        'capital-loss': capital_loss,
        'hours-per-week': hours_per_week,
        'native-country': native_country
    }

    if st.button("🚀 Predict"):
        with st.spinner("Analyzing..."):
            try:
                processed = preprocess_input(input_data)
                pred = model.predict(processed)[0]
                proba = model.predict_proba(processed)[0]
                label = ">50K" if pred == 1 else "≤50K"
                emoji = "💰" if pred == 1 else "💵"
                st.success(f"{emoji} Prediction: **{label}** with {max(proba)*100:.1f}% confidence")

                fig = px.bar(
                    x=["≤50K", ">50K"], y=proba,
                    labels={'x': 'Income Class', 'y': 'Probability'},
                    color=proba, color_continuous_scale='viridis',
                    title="Prediction Probability"
                )
                st.plotly_chart(fig, use_container_width=True)
            except Exception as e:
                st.error(f"Prediction error: {e}")

with tab2:
    st.markdown("### 📂 Batch Prediction (CSV Upload)")

    uploaded = st.file_uploader("Upload CSV with correct format", type=["csv"])
    required_cols = [
        'age', 'workclass', 'fnlwgt', 'educational-num', 'marital-status',
        'occupation', 'relationship', 'race', 'gender',
        'capital-gain', 'capital-loss', 'hours-per-week', 'native-country'
    ]

    if uploaded:
        try:
            df = pd.read_csv(uploaded)

            if all(col in df.columns for col in required_cols):
                if st.button("Run Batch Prediction"):
                    for col, mapping in [
                        ('workclass', workclass_map),
                        ('marital-status', marital_map),
                        ('occupation', occupation_map),
                        ('relationship', relationship_map),
                        ('race', race_map),
                        ('gender', gender_map),
                        ('native-country', country_map)
                    ]:
                        df[col] = df[col].map(mapping).fillna(0)

                    preds = model.predict(df[required_cols])
                    df['Prediction'] = ['>50K' if p == 1 else '≤50K' for p in preds]

                    st.success(f"✅ Processed {len(df)} records")
                    st.dataframe(df.head())

                    csv_out = df.to_csv(index=False).encode('utf-8')
                    st.download_button("Download CSV", csv_out, "predictions.csv", "text/csv")
            else:
                st.error("Uploaded file is missing one or more required columns.")
        except Exception as e:
            st.error(f"Error reading file: {e}")

with tab3:
    st.markdown("### ℹ️ Model Information")
    col1, col2 = st.columns(2)

    with col1:
        st.info("""
        - **Model**: Random Forest Classifier
        - **Training Accuracy**: 85.3%
        - **Dataset**: UCI Adult Dataset
        - **Features Used**: 13
        - **Target**: Income >50K or ≤50K
        """)

    with col2:
        features = [
            'Age', 'Education Level', 'Capital Gain', 'Hours/Week',
            'Occupation', 'Marital Status', 'Relationship', 'Work Class',
            'Race', 'Gender', 'Country', 'Capital Loss', 'Final Weight'
        ]
        importance = [0.18, 0.15, 0.14, 0.12, 0.10, 0.08,
                     0.07, 0.06, 0.04, 0.03, 0.02, 0.01, 0.0]

        fig = px.bar(
            x=importance, y=features, orientation='h',
            title="Feature Importance", labels={'x': 'Importance'},
            color=importance, color_continuous_scale='viridis'
        )
        st.plotly_chart(fig, use_container_width=True)

st.markdown("---")
st.markdown("<center>💡 Income Prediction System | Powered by UCI Dataset</center>", unsafe_allow_html=True)
```
